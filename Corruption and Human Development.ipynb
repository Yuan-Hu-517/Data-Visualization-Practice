```r
install.packages('ggplot2')
install.packages('ggthemes')
library(ggplot2)
library(ggthemes)
```

```r
df = read.csv('data.csv')
```

```r
pointsToLabel <- c("Russia", "Venezuela", "Iraq", "Myanmar", "Sudan",
                   "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                   "India", "Italy", "China", "South Africa", "Spane",
                   "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                   "United States", "Germany", "Britain", "Barbados", "Norway", "Japan",
                   "New Zealand", "Singapore")
```

```r
ggplot(df, aes(x = CPI, y = HDI)) +
  geom_point(aes(color = factor(Region)), size = 3, shape = 1) +
  geom_smooth(aes(group = 1), method = 'lm', formula = y~log(x), se = FALSE, color = 'red') +
  geom_text(aes(label = Country), color = "gray20", data = subset(df, Country %in% pointsToLabel),check_overlap = TRUE) +
  theme_bw() +
  scale_x_continuous(limits = c(0,11), breaks = 1:10, name = 'Corruption Perceptions Index, 2011 (10 = least corrupt)') +
  scale_y_continuous(limits = c(0.2,1.0), breaks = seq(0,1.0,0.1), name = 'Human Development Index, 2011 (1=Best)') +
  ggtitle("Corruption and Human development") +
  theme_economist_white() +
  theme(plot.title = element_text(hjust = 0.5))
```
